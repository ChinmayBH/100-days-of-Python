## Classes in python

```python
"""

A class in Python is a blueprint for creating objects.
It allows you to encapsulate data (attributes) and behavior (methods) into a single entity,
enabling the principles of object-oriented programming (OOP), such as encapsulation, inheritance, and polymorphism.

"""
```

```python
class Dog:
    """A class to represent a dog."""
    def __init__(self, name, breed):
        self.name = name      # Instance attribute
        self.breed = breed    # Instance attribute

# Create an object of the Dog class
my_dog = Dog("Buddy", "Golden Retriever")

# Access attributes
print(my_dog.name)   # Output: Buddy
print(my_dog.breed)  # Output: Golden Retriever
```

```python
class employee:
    def __init__(self, employee_id, position):
        self.employee_id = employee_id
        self.position = position

first_employee = employee(3118, "Software engineer")

print(first_employee.employee_id)
print(first_employee.position)
```

```python
# problem 
class student():
    def __init__(self,name,marks):
        self.name = name
        self.marks = marks

    def calculate_avarage(self):
        size = len(self.marks)
        total = 0
        for i in self.marks:
            total +=i
        avg = total/size
        return avg
        
    
```

```python
std1 = student("Chinmay",[2,3,4,5,6])
std1.calculate_avarage()
```

```python
class BankAccount():
    def __init__(self,holder_name, current_balance):
        self.holder_name = holder_name
        self.current_balance = current_balance

    def deposit_money(self,money):
        return f"your updated balance is {self.current_balance + money}"

    def widraw_money(self,windraw_money):
        if self.current_balance < windraw_money:
            return "your balance is not sufficient"
        else:
            return f"debited money, current balance is {self.current_balance  - windraw_money}"

    def Check_balance(self):
        return f"Your balance is {self.current_balance}"
        
```

```python
chinmay_bank_account = BankAccount("Chinmay",10000)
chinmay_bank_account.deposit_money(4000)
chinmay_bank_account.widraw_money(0)
chinmay_bank_account.Check_balance()
```
